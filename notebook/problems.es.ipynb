## Ejercicio 1 - Gráfico de líneas simple

Supongamos que, en tu oficina, te toma 10 segundos recorrer 25 metros caminando tranquilamente hacia la máquina de café. Si queremos expresar esta situación como una función lineal de la distancia respecto al tiempo, usamos la fórmula:

$
\text{distancia} = \text{velocidad} \times \text{tiempo}
$



Dado que la velocidad es constante, esta representa la pendiente de la función y se puede interpretar como la primera derivada de la distancia con respecto al tiempo. Es decir, nos indica qué tan rápido cambia la distancia mientras pasa el tiempo.

- **Calcula la velocidad constante a la que caminas:** Sabemos que recorriste 25 metros en 10 segundos. ¿Cuál es la velocidad (en metros/segundo)?
- **Construye la función lineal que describe cómo cambia la distancia con el tiempo:** Usa la fórmula `distancia = velocidad × tiempo` para calcular la distancia recorrida en cada segundo desde 0 a 10.
- **Usa Matplotlib para graficar cómo cambia la distancia con el tiempo:** Agrega etiquetas a los ejes y un título al gráfico.

```python
# x = np.arange(0, 11) # array con valores de 0 a 10

import numpy as np
import matplotlib.pyplot as plt

# Definir la velocidad
velocidad = 2.5  # metros por segundo

# Crear el array de tiempo (valores desde 0 a 10)
x = np.arange(0, 11)  # Tiempo en segundos

# Calcular las distancias correspondientes
distancia = velocidad * x  # Distancia recorrida en cada instante de tiempo

# Graficar la función lineal
plt.plot(x, distancia, color='red', marker='o', linestyle='-', label='Distancia vs Tiempo')

# Añadir título y etiquetas a los ejes
plt.title('Distancia recorrida en función del tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Distancia (m)')
plt.legend()

# Mostrar el gráfico
plt.grid(True)
plt.show()
```

## Ejercicio 2 - Representar visualmente datos categóricos.

Imagina que estás analizando cuántos estudiantes están matriculados en distintas carreras dentro de una universidad. Estos datos, que representan **categorías** (las carreras), son ideales para visualizar usando un **gráfico de barras**.

A continuación, se presenta un pequeño conjunto de datos con la cantidad de estudiantes por carrera.

```python

import pandas as pd
import matplotlib.pyplot as plt

# Crear el DataFrame
df_carreras = pd.DataFrame({
    'Carrera': ['Ingeniería', 'Medicina', 'Derecho', 'Psicología'],
    'Estudiantes': [120, 80, 60, 100]
})

# Visualizar los datos para comprobar
print(df_carreras)

# Crear el gráfico de barras
plt.bar(df_carreras['Carrera'], df_carreras['Estudiantes'], color='skyblue', edgecolor='black')

# Añadir título y etiquetas
plt.title('Cantidad de estudiantes por carrera')
plt.xlabel('Carrera')
plt.ylabel('Cantidad de Estudiantes')

# Rotar etiquetas del eje X si es necesario
plt.xticks(rotation=45)

# Mostrar el gráfico
plt.tight_layout()  # Ajustar el gráfico para que todo sea visible
plt.show()
```

1. Representa gráficamente cuántos estudiantes hay en cada carrera y utiliza un gráfico de barras vertical con Matplotlib. 
2. Etiqueta los ejes correctamente y agrega un título descriptivo.
3. Usa un color que ayude a destacar las diferencias entre categorías.



```python
import matplotlib.pyplot as plt

# Datos ficticios: número de estudiantes por carrera
carreras = ['Ingeniería', 'Derecho', 'Medicina', 'Arquitectura', 'Economía']
estudiantes = [120, 85, 150, 65, 95]

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))  # Tamaño del gráfico
plt.bar(carreras, estudiantes, color='royalblue', edgecolor='black')

# Etiquetas de los ejes y título
plt.xlabel("Carreras", fontsize=14)
plt.ylabel("Número de Estudiantes", fontsize=14)
plt.title("Distribución de Estudiantes por Carrera", fontsize=16)

# Mejoras visuales
plt.xticks(fontsize=12)  # Tamaño de las etiquetas del eje x
plt.yticks(fontsize=12)  # Tamaño de las etiquetas del eje y
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Líneas de cuadrícula en el eje Y

# Mostrar el gráfico
plt.tight_layout()  # Evitar superposiciones
plt.show()
```

Una excelente forma de **visualizar la relación entre dos variables numéricas** es mediante un **gráfico de dispersión** (`scatter plot`).

## Ejercicio 5 - Visualización de distribuciones con histogramas

En muchos casos, no solo nos interesa el valor de una variable, sino **cómo se distribuyen** esos valores.

Los histogramas son una herramienta visual muy útil para:

- Ver si los datos están **centrados** o dispersos.
- Detectar **asimetrías** (por ejemplo, si hay más valores bajos o altos).
- Identificar posibles **valores atípicos** (*outliers*).
- Observar la **forma general** de la distribución (normal, sesgada, bimodal, etc.).

Vamos a explorar cómo se distribuye el **largo del pétalo** (`petal_length`) en un conjunto de datos de flores.

```python
# df_iris.head() # Mostrar DataFrame

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Crear un DataFrame ficticio (reemplázalo con tus datos reales si los tienes)
data = {
    'petal_length': [1.4, 1.5, 1.3, 4.5, 4.7, 5.1, 1.6, 4.4, 5.2, 5.5, 1.7, 4.8, 5.0]
}
df_flores = pd.DataFrame(data)

# Crear el histograma
plt.figure(figsize=(8, 6))  # Tamaño del gráfico
sns.histplot(
    data=df_flores,
    x='petal_length',
    kde=True,  # Agregar línea de densidad para observar la forma general
    color='skyblue',  # Usar un color agradable
    bins=10  # Dividir los valores en 10 intervalos
)

# Etiquetas y título
plt.xlabel("Largo del Pétalo (cm)", fontsize=14)
plt.ylabel("Frecuencia", fontsize=14)
plt.title("Distribución del Largo del Pétalo", fontsize=16)
plt.grid(alpha=0.3)  # Cuadrícula ligera para claridad

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

1. Crea un **histograma con `Seaborn`** para visualizar la distribución del largo del pétalo (`petal_length`).
2. Ajusta el número de *bins* para tener mayor o menor nivel de detalle.
3. Agrega título, etiquetas y un estilo visual claro.

```python
# tu código

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Crear un DataFrame ficticio (puedes usar tus datos reales)
data = {
    'petal_length': [1.4, 1.5, 1.3, 4.5, 4.7, 5.1, 1.6, 4.4, 5.2, 5.5, 1.7, 4.8, 5.0]
}
df_flores = pd.DataFrame(data)

# Crear el histograma con Seaborn
plt.figure(figsize=(8, 6))  # Ajustar el tamaño del gráfico
sns.histplot(
    data=df_flores,
    x='petal_length',
    bins=15,  # Ajustar el nivel de detalle (mayor número de bins para más precisión)
    kde=True,  # Línea de densidad para visualizar la forma general
    color='orchid'  # Usar un color atractivo para el gráfico
)

# Agregar título y etiquetas
plt.title("Distribución del Largo del Pétalo", fontsize=16)
plt.xlabel("Largo del Pétalo (cm)", fontsize=14)
plt.ylabel("Frecuencia", fontsize=14)
plt.grid(alpha=0.3)  # Cuadrícula ligera para ayudar a la interpretación

# Mostrar el gráfico
plt.tight_layout()  # Asegurar que el diseño sea limpio
plt.show()
```

## Comparación por especie (distribuciones múltiples)

Ahora, compara la distribución del largo del pétalo para cada especie de flor usando `Seaborn` y el argumento `hue`.

Esto te permitirá analizar si hay **diferencias significativas** entre los grupos.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Crear un DataFrame ficticio (reemplázalo con tus datos reales)
data = {
    'petal_length': [1.4, 1.5, 1.3, 4.5, 4.7, 5.1, 1.6, 4.4, 5.2, 5.5, 1.7, 4.8, 5.0],
    'species': ['Setosa', 'Setosa', 'Setosa', 'Versicolor', 'Versicolor', 'Virginica', 'Setosa', 
                'Versicolor', 'Virginica', 'Virginica', 'Setosa', 'Versicolor', 'Virginica']
}
df_flores = pd.DataFrame(data)

# Crear el gráfico de comparación de distribuciones
plt.figure(figsize=(10, 6))  # Ajustar tamaño del gráfico
sns.histplot(
    data=df_flores,
    x='petal_length',
    hue='species',  # Diferenciar por especie con colores
    kde=True,  # Agregar línea de densidad para observar diferencias en las distribuciones
    palette='pastel',  # Usar una paleta clara y agradable
    bins=10,  # Número de intervalos
    multiple='stack'  # Apilar las distribuciones para una mejor comparación visual
)

# Etiquetas y título
plt.title("Distribución del Largo del Pétalo por Especie", fontsize=16)
plt.xlabel("Largo del Pétalo (cm)", fontsize=14)
plt.ylabel("Frecuencia", fontsize=14)
plt.legend(title="Especie", fontsize=12)  # Leyenda clara y con título
plt.grid(alpha=0.3)  # Cuadrícula ligera para mayor claridad

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

## Ejercicio 6 - Visualización de relaciones con mapas de calor (heatmap)

En ciencia de datos es importante comprender cómo se relacionan las variables entre sí. Una forma visual de hacerlo es usando mapas de calor, especialmente sobre matrices de correlación.

En este ejercicio vamos a visualizar cómo se relacionan las variables numéricas del conjunto `df_iris`.



1. Crea un mapa de calor que muestre la correlación entre las variables numéricas del DataFrame `df_iris`.Primero necesitas calcular la matriz de correlación usando `.corr()`.
2. Usa la librería **seaborn** para graficar el heatmap.
3. Activa la visualización de los valores dentro del gráfico `(annot=True)`.

```python
# df_iris

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Crear un DataFrame ficticio similar a df_iris (reemplázalo con tus datos reales)
data = {
    'sepal_length': [5.1, 4.9, 4.7, 6.3, 6.5, 6.7, 5.8],
    'sepal_width': [3.5, 3.0, 3.2, 3.3, 3.0, 3.1, 3.2],
    'petal_length': [1.4, 1.5, 1.3, 4.5, 4.7, 5.1, 5.2],
    'petal_width': [0.2, 0.2, 0.1, 1.5, 1.6, 2.0, 1.8]
}
df_iris = pd.DataFrame(data)

# Calcular la matriz de correlación
correlacion = df_iris.corr()

# Crear el mapa de calor
plt.figure(figsize=(8, 6))  # Ajustar tamaño del gráfico
sns.heatmap(
    correlacion,
    annot=True,  # Mostrar los valores dentro del mapa
    fmt='.2f',  # Formatear los valores numéricos con dos decimales
    cmap='coolwarm',  # Usar una paleta de colores clara
    linewidths=0.5,  # Separar las celdas con líneas
    square=True  # Mantener celdas cuadradas
)

# Agregar título
plt.title("Mapa de Calor - Correlación entre Variables Numéricas", fontsize=16)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

## Ejercicio 7 - Visualización de la distribución por categorías con boxplot

El boxplot (o diagrama de caja) permite comparar cómo se distribuye una variable numérica según distintas categorías.

Vamos a comparar la distribución del largo del pétalo (petal_length) entre las distintas especies de flores del conjunto `df_iris`. La idea es que grafiques un boxplot que compare el largo del pétalo (`petal_length`) entre las diferentes especies (species).


1. Usa `seaborn.boxplot()` indicando el eje `x` con la especie y `y` con el largo del pétalo.
2. Es recomendable usar plt.figure() para ajustar el tamaño del gráfico si es necesario.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Crear un DataFrame ficticio (puedes reemplazarlo con tus datos reales)
data = {
    'petal_length': [1.4, 1.5, 1.3, 4.5, 4.7, 5.1, 5.2, 1.6, 4.4, 5.2, 5.5, 1.7, 4.8, 5.0],
    'species': ['Setosa', 'Setosa', 'Setosa', 'Versicolor', 'Versicolor', 'Virginica', 'Virginica',
                'Setosa', 'Versicolor', 'Virginica', 'Virginica', 'Setosa', 'Versicolor', 'Virginica']
}
df_iris = pd.DataFrame(data)

# Crear el gráfico de boxplot
plt.figure(figsize=(8, 6))  # Ajustar el tamaño del gráfico
sns.boxplot(
    data=df_iris,
    x='species',
    y='petal_length',
    palette='muted'  # Usar una paleta de colores agradable
)

# Etiquetas y título
plt.xlabel("Especie", fontsize=14)
plt.ylabel("Largo del Pétalo (cm)", fontsize=14)
plt.title("Distribución del Largo del Pétalo por Especie", fontsize=16)

# Mostrar el gráfico
plt.tight_layout()  # Asegurar un diseño limpio
plt.show()




```
